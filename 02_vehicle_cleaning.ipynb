```python
import pandas as pd
import os

folder_path = r'C:\Users\abhishek\OneDrive\Documents\Bangalore-B2B-Intelligence-Engine\Data\Raw Data'

print("Starting the Vehicle Refinery Pipeline...\n")

try:
    all_files = os.listdir(folder_path)
    excel_files = [f for f in all_files if f.endswith('.xlsx') and 'vehicle' in f.lower()]
    
    if len(excel_files) == 0:
        print("Error: Could not find the transport vehicles file.")
    else:
        file_name = excel_files[0]
        file_path = os.path.join(folder_path, file_name)
        print(f"Success! Loaded file: {file_name}")
        
        df_vehicles = pd.read_excel(file_path, engine='openpyxl')
        df_vehicles = df_vehicles.loc[:, ~df_vehicles.columns.str.contains('^Unnamed')]
        
        print(f"Original Vehicle Count: {len(df_vehicles)}")
        
        print("\n--- MESSY MODELS BEFORE CLEANING ---")
        print(df_vehicles['MODEL'].value_counts())
        
        print("\n--- MESSY LOCATIONS BEFORE CLEANING ---")
        print(df_vehicles['LOCATION'].value_counts())
        
except Exception as e:
    print(f"A system error occurred: {e}")
```

```python
print("\n--- INITIATING CLEANING PROTOCOL ---")
model_corrections = {
    'ape': 'Piaggio Ape',
    'pe': 'Piaggio Ape',
    'Baja re': 'Bajaj RE',
    'Baja Re': 'Bajaj RE',
    'Bajaj Re': 'Bajaj RE',
    'Bajaj RE': 'Bajaj RE',
    'Bajaj CNG RE': 'Bajaj RE CNG',
    'Bajaj RE CNG': 'Bajaj RE CNG',
    'Dost LS': 'Ashok Leyland Dost',
    'Ashok Leyland Dost': 'Ashok Leyland Dost',
    'Ashok Leyland Dost CNG': 'Ashok Leyland Dost CNG',
    'Mahindra supro': 'Mahindra Supro',
    'TVS King': 'TVS King'
}

df_vehicles['MODEL'] = df_vehicles['MODEL'].str.strip().replace(model_corrections)
df_vehicles['LOCATION'] = df_vehicles['LOCATION'].str.strip().str.title()
df_vehicles['LOCATION'] = df_vehicles['LOCATION'].replace('Yeshwanthpura', 'Yeshwanthpur')

print("Spelling mistakes neutralized!")
print("\n--- PERFECTED MODELS ---")
print(df_vehicles['MODEL'].value_counts())

output_folder = r'C:\Users\abhishek\OneDrive\Documents\Bangalore-B2B-Intelligence-Engine\Data\Processed'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

output_file = os.path.join(output_folder, 'clean_vehicles.csv')
df_vehicles.to_csv(output_file, index=False)

print(f"\n permanently saved to:")
print(output_file)
```
